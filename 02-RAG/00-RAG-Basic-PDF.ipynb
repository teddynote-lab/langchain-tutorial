ê° ë¬¸ì„œ í˜ì´ì§€ì—ëŠ” **ë©”íƒ€ë°ì´í„°(Metadata)** ê°€ í•¨ê»˜ ì €ì¥ë©ë‹ˆë‹¤. ì´ëŠ” **ë„ì„œê´€ ë„ì„œì¹´ë“œ**ì™€ ê°™ì€ ì—­í• ë¡œ, ì¶œì²˜ ì •ë³´ë¥¼ ë‹´ê³  ìˆì–´ ë‚˜ì¤‘ì— ë‹µë³€ì˜ ê·¼ê±°ë¥¼ ì œì‹œí•  ë•Œ ìœ ìš©í•©ë‹ˆë‹¤:

````python
# 10ë²ˆì§¸ í˜ì´ì§€ ë¬¸ì„œì˜ ì „ì²´ ì •ë³´(ë©”íƒ€ë°ì´í„° í¬í•¨)ë¥¼ ë”•ì…”ë„ˆë¦¬ í˜•íƒœë¡œ ì¶œë ¥
# page_content: ì‹¤ì œ í…ìŠ¤íŠ¸ ë‚´ìš©, metadata: íŒŒì¼ëª…, í˜ì´ì§€ ë²ˆí˜¸ ë“± ë¶€ê°€ ì •ë³´
docs[10].__dict__
````

````python
# ë‹¨ê³„ 2: ë¬¸ì„œ ë¶„í• (Split Documents)
# ê¸´ ë¬¸ì„œë¥¼ ê²€ìƒ‰í•˜ê¸° ì ì ˆí•œ í¬ê¸°ë¡œ ë‚˜ëˆ„ëŠ” í…ìŠ¤íŠ¸ ìŠ¤í”Œë¦¬í„° ìƒì„±
# chunk_size=500: ê° ì²­í¬ì˜ ìµœëŒ€ ê¸€ì ìˆ˜
# chunk_overlap=50: ì¸ì ‘í•œ ì²­í¬ ê°„ ê²¹ì¹˜ëŠ” ê¸€ì ìˆ˜ (ë¬¸ë§¥ ìœ ì§€ë¥¼ ìœ„í•¨)
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
# ë¡œë“œëœ ë¬¸ì„œë“¤ì„ ì„¤ì •ëœ í¬ê¸°ë¡œ ë¶„í• 
split_documents = text_splitter.split_documents(docs)
print(f"ë¶„í• ëœ ì²­í¬ì˜ìˆ˜: {len(split_documents)}")  # ë¶„í• ëœ ì´ ì²­í¬ ê°œìˆ˜ ì¶œë ¥
````

````python
# ë‹¨ê³„ 3: ì„ë² ë”©(Embedding) ìƒì„±
# í…ìŠ¤íŠ¸ë¥¼ ë²¡í„°(ìˆ«ì ë°°ì—´)ë¡œ ë³€í™˜í•˜ëŠ” ì„ë² ë”© ëª¨ë¸ ìƒì„±
# text-embedding-3-small: OpenAIì˜ ìµœì‹  ì„ë² ë”© ëª¨ë¸ (ì„±ëŠ¥ ì¢‹ê³  ë¹„ìš© íš¨ìœ¨ì )
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
````

````python
# ë‹¨ê³„ 4: DB ìƒì„±(Create DB) ë° ì €ì¥
# ë¶„í• ëœ ë¬¸ì„œë“¤ì„ ì„ë² ë”©ìœ¼ë¡œ ë³€í™˜í•˜ì—¬ FAISS ë²¡í„° ë°ì´í„°ë² ì´ìŠ¤ì— ì €ì¥
# FAISS: Facebookì´ ê°œë°œí•œ ê³ ì† ìœ ì‚¬ë„ ê²€ìƒ‰ ë¼ì´ë¸ŒëŸ¬ë¦¬
vectorstore = FAISS.from_documents(documents=split_documents, embedding=embeddings)
````

````python
# ë²¡í„° ë°ì´í„°ë² ì´ìŠ¤ì—ì„œ "êµ¬ê¸€"ê³¼ ìœ ì‚¬í•œ ë¬¸ì„œ ê²€ìƒ‰ í…ŒìŠ¤íŠ¸
# similarity_search: ì¿¼ë¦¬ì™€ ê°€ì¥ ìœ ì‚¬í•œ ë¬¸ì„œë“¤ì„ ë°˜í™˜ (ê¸°ë³¸ 4ê°œ)
for doc in vectorstore.similarity_search("êµ¬ê¸€"):
    print(doc.page_content)  # ê²€ìƒ‰ëœ ê° ë¬¸ì„œì˜ ë‚´ìš© ì¶œë ¥
````

````python
# ë‹¨ê³„ 5: ê²€ìƒ‰ê¸°(Retriever) ìƒì„±
# ë²¡í„° ë°ì´í„°ë² ì´ìŠ¤ë¥¼ ê²€ìƒ‰ ê°€ëŠ¥í•œ retrieverë¡œ ë³€í™˜
# retrieverëŠ” ì¿¼ë¦¬ë¥¼ ë°›ì•„ ê´€ë ¨ ë¬¸ì„œë¥¼ ìë™ìœ¼ë¡œ ì°¾ì•„ì£¼ëŠ” ì—­í• 
retriever = vectorstore.as_retriever()
````

### ğŸ” ê²€ìƒ‰ê¸° í…ŒìŠ¤íŠ¸

ìƒì„±ëœ ê²€ìƒ‰ê¸°ê°€ ì œëŒ€ë¡œ ì‘ë™í•˜ëŠ”ì§€ **ì‹¤ì œ ì§ˆë¬¸**ìœ¼ë¡œ í…ŒìŠ¤íŠ¸í•´ë´…ì‹œë‹¤:

````python
# ê²€ìƒ‰ê¸°ì— ì§ˆë¬¸ì„ ë„£ì–´ì„œ ê´€ë ¨ ë¬¸ì„œ ì²­í¬ë“¤ì„ ê²€ìƒ‰
# invoke: ê²€ìƒ‰ê¸°ë¥¼ ì‹¤í–‰í•˜ëŠ” ë©”ì„œë“œ
# ì§ˆë¬¸ê³¼ ê´€ë ¨ëœ ë¬¸ì„œë“¤ì„ ìœ ì‚¬ë„ ìˆœìœ¼ë¡œ ë°˜í™˜
retriever.invoke("ì‚¼ì„±ì „ìê°€ ìì²´ ê°œë°œí•œ AI ì˜ ì´ë¦„ì€?")
````

````python
# ë‹¨ê³„ 6: í”„ë¡¬í”„íŠ¸ ìƒì„±(Create Prompt)  
# AIì—ê²Œ ê²€ìƒ‰ëœ ë¬¸ì„œë¥¼ ë°”íƒ•ìœ¼ë¡œ ì§ˆë¬¸ì— ë‹µë³€í•˜ë„ë¡ ì§€ì‹œí•˜ëŠ” í”„ë¡¬í”„íŠ¸ í…œí”Œë¦¿
prompt = PromptTemplate.from_template(
    """You are an expert assistant for question-answering tasks. 
Use the following pieces of retrieved context to answer the question accurately and comprehensively.

Instructions:
- Only use information from the provided context
- If you don't know the answer based on the context, say "ì£„ì†¡í•˜ì§€ë§Œ ì œê³µëœ ë¬¸ì„œì—ì„œëŠ” í•´ë‹¹ ì •ë³´ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤."
- Answer in Korean with a clear and professional tone
- Cite specific parts of the context when possible

#Context: 
{context}

#Question:
{question}

#Answer:"""
)
````

````python
# ë‹¨ê³„ 7: ì–¸ì–´ëª¨ë¸(LLM) ìƒì„±
# ì§ˆë¬¸-ë‹µë³€ì„ ìˆ˜í–‰í•  ChatGPT ëª¨ë¸ ì„¤ì •
# gpt-4.1: ìµœì‹  GPT-4 ëª¨ë¸, temperature=0: ì¼ê´€ëœ ë‹µë³€ ìƒì„± (ì°½ì˜ì„± ìµœì†Œí™”)
llm = ChatOpenAI(model_name="gpt-4o", temperature=0)
````

````python
# ë‹¨ê³„ 8: ì²´ì¸(Chain) ìƒì„±
# ê²€ìƒ‰â†’í”„ë¡¬í”„íŠ¸â†’LLMâ†’ì¶œë ¥ íŒŒì‹±ì˜ ì „ì²´ íŒŒì´í”„ë¼ì¸ì„ í•˜ë‚˜ì˜ ì²´ì¸ìœ¼ë¡œ ì—°ê²°
# {context: retriever, question: RunnablePassthrough()}: ì§ˆë¬¸ì„ ë°›ì•„ ê²€ìƒ‰ ê²°ê³¼ì™€ í•¨ê»˜ ì „ë‹¬
# |: íŒŒì´í”„ë¼ì´í”„ ì—°ì‚°ìë¡œ ê° ë‹¨ê³„ë¥¼ ìˆœì°¨ì ìœ¼ë¡œ ì—°ê²°
chain = (
    {"context": retriever, "question": RunnablePassthrough()}  # ì§ˆë¬¸ì„ ê²€ìƒ‰ê¸°ì— ì „ë‹¬í•˜ê³  ê²°ê³¼ë¥¼ contextë¡œ ì„¤ì •
    | prompt  # ê²€ìƒ‰ ê²°ê³¼ì™€ ì§ˆë¬¸ì„ í”„ë¡¬í”„íŠ¸ í…œí”Œë¦¿ì— ì‚½ì…
    | llm  # ì™„ì„±ëœ í”„ë¡¬í”„íŠ¸ë¥¼ LLMì— ì „ë‹¬í•˜ì—¬ ë‹µë³€ ìƒì„±
    | StrOutputParser()  # LLM ì‘ë‹µì„ ë¬¸ìì—´ë¡œ ë³€í™˜
)
````

### ğŸ¯ RAG ì‹œìŠ¤í…œ ì‹¤í–‰í•˜ê¸°

ë“œë””ì–´ **ì™„ì„±ëœ RAG ì²´ì¸**ì„ ì‹¤í–‰í•  ì‹œê°„ì…ë‹ˆë‹¤! ğŸš€

ì´ì œ ì‚¬ìš©ìê°€ ì§ˆë¬¸ì„ ì…ë ¥í•˜ë©´:
1. **ğŸ” ê²€ìƒ‰ê¸°**ê°€ ê´€ë ¨ ë¬¸ì„œë¥¼ ì°¾ê³ 
2. **ğŸ“ í”„ë¡¬í”„íŠ¸**ê°€ ì§ˆë¬¸ê³¼ ë¬¸ì„œë¥¼ ì¡°í•©í•˜ì—¬
3. **ğŸ¤– AI**ê°€ ë‹µë³€ì„ ìƒì„±í•©ë‹ˆë‹¤

ëª¨ë“  ê³¼ì •ì´ **ìë™ìœ¼ë¡œ** ì‹¤í–‰ë©ë‹ˆë‹¤:

````python
# ì²´ì¸ ì‹¤í–‰(Run Chain)
# ì™„ì„±ëœ RAG íŒŒì´í”„ë¼ì¸ì— ì§ˆë¬¸ì„ ì…ë ¥í•˜ì—¬ ì‹¤í–‰
question = "ì‚¼ì„±ì „ìê°€ ìì²´ ê°œë°œí•œ AI ì˜ ì´ë¦„ì€?"
# invoke: ì²´ì¸ì„ ì‹¤í–‰í•˜ëŠ” ë©”ì„œë“œ (ì§ˆë¬¸ â†’ ê²€ìƒ‰ â†’ í”„ë¡¬í”„íŠ¸ â†’ LLM â†’ ë‹µë³€)
response = chain.invoke(question)
print(response)  # ìµœì¢… ë‹µë³€ ì¶œë ¥
````

# ğŸ”§ ì „ì²´ ì½”ë“œ í†µí•©

## ğŸ’¡ ì™„ì„±ëœ RAG ì‹œìŠ¤í…œ

ì§€ê¸ˆê¹Œì§€ ë‹¨ê³„ë³„ë¡œ êµ¬í˜„í•œ ë‚´ìš©ì„ **í•˜ë‚˜ì˜ ì™„ì „í•œ ì½”ë“œ**ë¡œ í†µí•©í–ˆìŠµë‹ˆë‹¤! 

### ğŸ¯ ì´ ì½”ë“œì˜ íŠ¹ì§•

- **ğŸ“– Production Ready**: ì‹¤ì œ í”„ë¡œì íŠ¸ì—ì„œ ë°”ë¡œ ì‚¬ìš© ê°€ëŠ¥
- **âš™ï¸ ìµœì í™”ëœ ì„¤ì •**: chunk_sizeë¥¼ 1000ìœ¼ë¡œ ì¦ê°€ì‹œì¼œ ë” ë§ì€ ë§¥ë½ í¬í•¨
- **ğŸ”„ ì¬ì‚¬ìš© ê°€ëŠ¥**: ë‹¤ë¥¸ PDF ë¬¸ì„œì—ë„ ë°”ë¡œ ì ìš© ê°€ëŠ¥
- **ğŸ“ ì™„ì „í•œ íŒŒì´í”„ë¼ì¸**: Importë¶€í„° ì‹¤í–‰ê¹Œì§€ ëª¨ë“  ê³¼ì • í¬í•¨

### ğŸš€ ì‹¤í–‰ ë°©ë²•

ì•„ë˜ ì½”ë“œë¥¼ **í•œ ë²ˆì— ì‹¤í–‰**í•˜ë©´ ì™„ì „í•œ RAG ì‹œìŠ¤í…œì´ êµ¬ë™ë©ë‹ˆë‹¤:

````python
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ import
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_community.vectorstores import FAISS
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI, OpenAIEmbeddings

# ë‹¨ê³„ 1: ë¬¸ì„œ ë¡œë“œ(Load Documents)
loader = PyMuPDFLoader("data/SPRI_AI_Brief_2023ë…„12ì›”í˜¸_F.pdf")
docs = loader.load()

# ë‹¨ê³„ 2: ë¬¸ì„œ ë¶„í• (Split Documents)  
# chunk_sizeë¥¼ 1000ìœ¼ë¡œ ì„¤ì •í•˜ì—¬ ë” ë§ì€ ë§¥ë½ ì •ë³´ í¬í•¨
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=50)
split_documents = text_splitter.split_documents(docs)

# ë‹¨ê³„ 3: ì„ë² ë”©(Embedding) ìƒì„±
# text-embedding-3-small ëª¨ë¸ ì‚¬ìš© (ì„±ëŠ¥ê³¼ ë¹„ìš©ì˜ ìµœì  ê· í˜•)
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")

# ë‹¨ê³„ 4: DB ìƒì„±(Create DB) ë° ì €ì¥
# FAISS ë²¡í„° ë°ì´í„°ë² ì´ìŠ¤ ìƒì„± ë° ë¬¸ì„œ ì €ì¥
vectorstore = FAISS.from_documents(documents=split_documents, embedding=embeddings)

# ë‹¨ê³„ 5: ê²€ìƒ‰ê¸°(Retriever) ìƒì„±
# ì§ˆì˜ì™€ ìœ ì‚¬í•œ ë¬¸ì„œë¥¼ ê²€ìƒ‰í•˜ëŠ” retriever ìƒì„±
retriever = vectorstore.as_retriever()

# ë‹¨ê³„ 6: í”„ë¡¬í”„íŠ¸ ìƒì„±(Create Prompt)
# ê°œì„ ëœ í”„ë¡¬í”„íŠ¸ í…œí”Œë¦¿ìœ¼ë¡œ ë” ì •í™•í•œ ë‹µë³€ ìœ ë„
prompt = PromptTemplate.from_template(
    """You are an expert assistant for question-answering tasks. 
Use the following pieces of retrieved context to answer the question accurately and comprehensively.

Instructions:
- Only use information from the provided context
- If you don't know the answer based on the context, say "ì£„ì†¡í•˜ì§€ë§Œ ì œê³µëœ ë¬¸ì„œì—ì„œëŠ” í•´ë‹¹ ì •ë³´ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤."
- Answer in Korean with a clear and professional tone
- Cite specific parts of the context when possible

#Context: 
{context}

#Question:
{question}

#Answer:"""
)

# ë‹¨ê³„ 7: ì–¸ì–´ëª¨ë¸(LLM) ìƒì„±
# GPT-4o ëª¨ë¸ ì‚¬ìš©, temperature=0ìœ¼ë¡œ ì¼ê´€ëœ ë‹µë³€ ìƒì„±
llm = ChatOpenAI(model_name="gpt-4o", temperature=0)

# ë‹¨ê³„ 8: ì²´ì¸(Chain) ìƒì„±  
# ì „ì²´ RAG íŒŒì´í”„ë¼ì¸ì„ í•˜ë‚˜ì˜ ì‹¤í–‰ ê°€ëŠ¥í•œ ì²´ì¸ìœ¼ë¡œ êµ¬ì„±
chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
````

````python
# ì²´ì¸ ì‹¤í–‰(Run Chain) - ìµœì¢… í…ŒìŠ¤íŠ¸
# RAG ì‹œìŠ¤í…œì— ì§ˆë¬¸ì„ ì…ë ¥í•˜ì—¬ ì™„ì „í•œ íŒŒì´í”„ë¼ì¸ ì‹¤í–‰
question = "ì‚¼ì„±ì „ìê°€ ìì²´ ê°œë°œí•œ AI ì˜ ì´ë¦„ì€?"
# ì§ˆë¬¸ â†’ ë¬¸ì„œê²€ìƒ‰ â†’ í”„ë¡¬í”„íŠ¸ìƒì„± â†’ AIë‹µë³€ â†’ ê²°ê³¼ë°˜í™˜ì˜ ì „ì²´ ê³¼ì •ì´ ìë™ ì‹¤í–‰
response = chain.invoke(question)
print(response)
````
